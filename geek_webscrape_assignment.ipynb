```python
import re
import requests
import pandas as pd
from bs4 import BeautifulSoup
url = 'https://www.geeksforgeeks.org/what-is-web-scraping-and-how-to-use-it/'
response= requests.get(url)
response
webpage_content = BeautifulSoup(response.text, 'html.parser')
#title extraction
title_pattern = r"<title>(.*?)</title>"
title_match = re.search(title_pattern, response.text)
webpage_title = title_match.group(1)
webpage_title
#Content titles extratction
titles = webpage_content.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6'])
titles_text = [title.get_text() for title in titles]
titles_text
#Links extraction
links = webpage_content.find_all('a')
links_href = [link.get('href') for link in links]
links_href
#Date extraction
date_pattern = r"<span>(\d{2} \w{3}, \d{4})</span>"
date_match = re.search(date_pattern, response.text)
date = date_match.group(1)
date
#Image extraction
images_pattern = r'<img src="(.*?)"'
images = re.findall(images_pattern, response.text)
images
```

```python
num_titles = len(titles_text)
num_images = len(images)

if num_images < num_titles:
    images.extend([None] * (num_titles - num_images)) 
elif num_images > num_titles:
    images = images[:num_titles]


data = {
    'Webpage Title': [webpage_title] * num_titles, 
    'Content Titles': titles_text,
    'Links': links_href[:num_titles],  
    'Date Last Updated': [date] * num_titles, 
    'Image Links': images
}

df = pd.DataFrame(data)

df.to_csv('webpage_data.csv', index=False)

print("Data saved to 'webpage_data.csv'")
```
